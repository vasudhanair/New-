```python
#dependencies
import pandas as pd
import json
import requests
from config import weather_api_key

base_url =  "http://api.openweathermap.org/data/2.5/weather?"
store_path = "Resources/directory.csv"
```

```python
#read CSV into a DF
store_data = pd.read_csv(store_path)
store_data.head()
```

```python
#created a DF with Starbucks only
starbucks_df = store_data[store_data["Brand"].isin(["Starbucks"])]
starbucks_df.head()
```

```python
#created a DF with US locations only
US_starbucks_df = starbucks_df[starbucks_df["Country"].isin(["US"])]
US_starbucks_df.head()
```

```python
#removed N/A
US_starbucks_df = US_starbucks_df.dropna(subset = ["Postcode"])

US_starbucks_df.head()
```

```python
#created a city list based on Starbucks locations
starbucks_cities = US_starbucks_df["City"].unique()

starbucks_cities
```

```python
# Loop through the list of cities and perform a request for data on each
api_key = weather_api_key
weather_data = []
for city in starbucks_cities:
    query_url = base_url + "appid=" + api_key + "&q=" + city + "&units=metric"
    city_info = requests.get(query_url).json()
    try:
        city_info = requests.get(query_url).json()
        city_name = city_info["name"]
        city_lat = city_info["coord"]["lat"]
        city_lng = city_info["coord"]["lon"]
        city_temp_max = city_info["main"]["temp_max"]
        city_humidity = city_info["main"]["humidity"]
        city_cloud = city_info["clouds"]["all"]
        city_wind_speed = city_info["wind"]["speed"]
        city_country = city_info["sys"]["country"]
        city_date = city_info["dt"]
        weather_data.append([city_name, city_lat, city_lng, city_temp_max, city_humidity, city_cloud, city_wind_speed, city_country, city_date])
        print(f"Weather found in {city_name}")

    except(KeyError):
        print("City not found")
print("Weather Data Complete")
```

```python
#created a weather DF and saved it into a CSV
weather_df = pd.DataFrame(weather_data, columns = ["City", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Country", "Date"])
weather_df.to_csv("Resources/weather_data_from_starbucks.csv")

weather_df.head()
```
